# **Project Summary -**

This project focuses on analyzing an automobile dataset to derive actionable insights that support a data-driven business model for an automotive company. The dataset includes 205 records and 26 attributes, covering various aspects of vehicle specifications, such as engine size, horsepower, fuel type, body style, and price. The primary objective is to use this information to optimize pricing, understand customer preferences, identify market segments, and enhance overall business performance.

The project began with comprehensive data cleaning and preprocessing. Several columns such as normalized-losses, bore, stroke, horsepower, peak-rpm, and price were stored as objects due to the presence of missing values or placeholders like "?". These columns were converted to appropriate numeric types after handling missing data using techniques like mean/mode imputation or removal of invalid entries. This step was critical to ensure the accuracy of further statistical analysis and model building.

Next, exploratory data analysis (EDA) was conducted to understand relationships among various features. Correlation matrices and scatter plots revealed that engine-size, curb-weight, and horsepower had strong positive correlations with price, indicating that more powerful and heavier cars tend to be priced higher. On the other hand, features like city-mpg and highway-mpg were negatively correlated with price, suggesting that fuel-efficient vehicles are generally more affordable. These insights laid the groundwork for developing pricing strategies based on vehicle characteristics.

To support strategic pricing and product planning, a linear regression model was built to predict vehicle price. The model considered significant predictors such as engine-size, horsepower, curb-weight, and body-style. The model performed well and provided interpretable coefficients, helping the company understand how different features impact pricing. This predictive capability can assist in setting optimal prices for new car models or evaluating the competitiveness of existing products.

In addition to pricing, clustering algorithms like K-Means were used to segment the vehicles into distinct market groups. Based on features such as fuel efficiency, engine size, and price, the models identified groups like economy cars, mid-range vehicles, and high-performance models. These segments help in tailoring marketing strategies and product designs to meet specific customer needs and preferences.

The analysis also explored fuel efficiency optimization by evaluating the trade-off between performance metrics (like horsepower) and mileage (city-mpg, highway-mpg). This helped identify configurations that offer a good balance of power and fuel economy, which is essential for targeting cost-conscious or environmentally aware consumers.

Furthermore, by analyzing variables such as symboling and normalized-losses, the project provides a foundation for risk assessment, which can be utilized in warranty planning or insurance offerings. High-risk models may be subject to different service packages or extended warranty terms.
